```python
df_accident_carac_ville=df_accident_carac[df_accident_carac['agg']==2]#on récupère les accidents ayant lieu dans une commune
```

```python
#on fusionne les bases avec les identifiants d'accident
df_accident=df_accident_usagers.merge(df_accident_carac_ville, on='Num_Acc')
```

```python
#on récupère les informations qui nous intéressent (voir site pour leur signification)
df_accident=df_accident[['Num_Acc', 'com', 'dep', 'adr', 'catu', 'grav', 'locp']]
```

```python
df_accident_metrop=df_accident[df_accident['dep']<100]#on récupère les informations pour la métropole
df_accident_metrop
```

```python
#on met une norme commune pour les identifiants de lieu : 2 chiffres pour le département et 3 pour la commune
def normalise_dep(x):
    if float(x)>=100:
        return str(x)[0]+str(x)[1]
    else :
        return str(0)+str(x)[0]
def normalise_com(x):
    if float(x)<100 :
        return '0'+str(round(float(x),0))[0:2]
    else : 
        return str(round(float(x),0))[0:3]
df_accident['dep']=df_accident['dep'].apply(normalise_dep)
df_accident['com']=df_accident['com'].apply(normalise_com)
df_accident
```

```python
#on continue en prenant en compte les départements 1 à 9 mal écrits
for nombre in range(1,10) :
    df_accident['com']=df_accident['com'].replace('0'+str(nombre)+'.', '00'+str(nombre))
df_accident['COD_MOD']=df_accident['dep'].apply(str)+df_accident['com'].apply(str)
```

```python
#on s'occupe du cas de la corse
df_communes=pd.read_csv(r'C:/Users/sylva/OneDrive/Bureau/barometre/bases complementaires/codecommunes.csv', sep=',')
df_communes['COD_MOD']=df_communes['com'].apply(str)
def prendre2(x):
    if str(x)=='A' or str(x)=='B':
        return str(0)
    else :
        return str(x)
def prendresansAB (x):
    return str(x)[0]+prendre2(str(x)[1])+str(x)[2:5]
df_communes['COD_MOD']=df_communes['COD_MOD'].apply(prendresansAB)
df_accident=df_accident.merge(df_communes[['COD_MOD','nccenr']], how='left')
df_accident
```

```python
df_accident['dep'].unique()
```

```python
#on récupère uniquement les accidents concernant des piétons
df_accident_pieton=df_accident[df_accident['catu']==3]
df_accident_pieton
```

```python
#on enlève le numéro d'arrondissement (pas pris en compte dans le 1er baromètre)
def sans_arrond (x):
    p=re.compile(r'Arrondissement')
    m=p.search(str(x))
    if m==None :
        return x
    else :
        return x.split()[0] 
df_accident_pieton['nom_ville']=df_accident_pieton['nccenr'].apply(sans_arrond)
```

```python
df_accident_pieton['Ville']=df_accident_pieton['nom_ville']+' '+df_accident_pieton['dep'].apply(str)
```

```python
#fonction pour vérifier si une ville n'a pas un problème de désignation (c'est le cas des villes Le ...)
import re
def estla(x):
    p=re.compile(r'ntib')
    m=p.search(str(x))
    if m==None :
        return False
    else :
        return True
df_accident_pieton[df_accident_pieton['nom_ville'].apply(estla)]
```

```python
#on compte le nombre d'accident par ville
liste_accident_ville=[]
liste_accident_nombre=[]
for ville in df_accident_pieton['Ville'].unique():
    liste_accident_nombre.append(df_accident_pieton[df_accident_pieton['Ville']==ville].count()[0])
    liste_accident_ville.append(ville)
liste_accident=pd.DataFrame(np.array(liste_accident_ville))
```

```python
liste_accident['Ville']=liste_accident
```

```python
#on crée un dataframe avec le nombre d'accidents piétons et la ville associée
liste_accident['nombre']=np.array(liste_accident_nombre)
df_accident=liste_accident[['nombre','Ville']]
df_accident=df_accident.rename(columns={'ville':'Ville', 'nombre':"nombre d'accidents pietons"})
df_accident
```

```python
#on exporte la base pour qu'elle soit réunie avec les autres dans préparation_tableau_baromètre
df_accident.to_csv(r'C:\Users\sylva\OneDrive\Bureau\barometre\bases complementaires\accidents.csv', index=False, sep=';')
```
